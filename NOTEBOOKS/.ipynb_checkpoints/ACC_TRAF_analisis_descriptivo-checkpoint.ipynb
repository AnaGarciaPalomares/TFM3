## ANÁLISIS ESTADÍSTICO DE LA SERIE TEMPORAL - ESTACIONARIEDAD

Debemos verificar la estacionariedad porque no solo facilita el modelado de series temporales, sino que es una **suposición subyacente en muchos métodos de pronóstico de series temporales**.

Las características principales de una serie temporal son:

- **Sus propiedades estadísticas no varían con el paso del tiempo** (la media, la varianza, y la autocorrelación permanecen constantes a lo largo del tiempo).

- Una serie temporal estacionaria **no tendrá tendencias ni patrones estacionales**.

**1. PROPIEDADES ESTADISTICAS EN EL TIEMPO**

```python
num_sin_año['2019'].describe()
```

**Autocorrelación**

```python
fig,ax = plt.subplots(2,1,figsize=(15,8))
fig = sm.graphics.tsa.plot_acf(num_sin_año, lags=200, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(num_sin_año, lags=200, ax=ax[1])
ax[0].set_xlim(0, 200)
ax[1].set_xlim(0, 200)
plt.show()
```

Los picos del ACF en los retardos 7, 14, 21, ... indican una estacionalidad de 7.

Además, el ACF que decae lentamente indica que hay una tendencia

**2. PRUEBA DE DICKEY FULLER**

Usaremos la prueba de **Dickey Fuller** para verificar la estacionariedad en nuestros datos:
    
    H0: no hay estacionariedad
    H1: hay estacionariedad
En esta prueba se evalúa la hipótesis nula, que confirma que existe una raíz unitaria en la serie temporal.

```python
adft = adfuller(num_sin_año,autolag='AIC')
```

```python
output_df = pd.DataFrame({'Values':[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']],\
                          'Metric':['Test Statistics','p-value','No. of lags used','Number of observations used',\
                            'critical value (1%)', 'critical value (5%)', 'critical value (10%)']})
print(output_df)
```

Nuestros datos son **estacionarios** porque nuestro **p-valor < 5%** y nuestro estadístico de prueba es menor que el valor crítico, es decir, estamos en la zona de rechazo de la hipotesis nula para un intervalo de confianza del 5%.

**3. PRUEBA DEL EXPONENTE DE HURST**

Método estadístico utilizado para medir la **memoria de largo plazo de las series temporales**: utiliza las autocorrelaciones de la serie temporal y el ratio al cual estas descienden a medida que el retardo entre pares de valores aumenta.

Es decir, nos permiten probar el grado en que los valores presentes cambian con los valores pasados. Si no hay estacionariedad en el conjunto de datos, un cambio en los valores presentes no provocará un cambio significativo en los valores pasados.

```python
H, c, val = compute_Hc(num_sin_año)
```

```python
print('Hurst exponent = {:.4f}'.format(H))
```

H < 0,5 --> proceso anti-persistente (reversión a la media)

Un valor de H < 0,5 indica una serie temporal con autocorrelación negativa, una observación positiva probablemente precederá una observación negativa y vice-versa. En este supuesto se asume que la serie temporal tiende a revertir a la media, y por ello, que lo adecuado probablemente sea diseñar un sistema de reversión a la media.

**4. DESCOMPOSION - Tendencia y componente cíclico**

Vamos a realizar una descomposición de la serie temporal (tendencia, estacionalidad y residuo).

```python
decompose = seasonal_decompose(num_sin_año['num_claims'],model='additive', period=30)
decompose.plot()
plt.show()
```

Otro método para descomponer la serie es por medio del **Filtro Hodrick-Prescott**: método para separar tendencia y componente ciclico.

```python
num_cycle, num_trend = sm.tsa.filters.hpfilter(num_sin_año['num_claims'])
```

```python
num_decomp = num_sin_año[['num_claims']].copy()
num_decomp['ciclico'] = num_cycle
num_decomp['tendencia'] = num_trend
```

```python
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111)
num_decomp[['num_claims','tendencia', 'ciclico']]['2019-01-01':'2021-12-31'].plot(ax=ax, fontsize=16)
legend = ax.get_legend()
legend.prop.set_size(20)
```

**Test Mann Kendall**
   
    H0: no hay tendencia
    H1: hay tendencia

Aunque ya vemos por la descomposión de la serie, que los datos recogen una tendencia, refozaremos nuestro análisis por medio del test Mann Kendall.

```python
mk.original_test(num_sin_año['num_claims'])
```

p-value < 0.05, hay tendencia en la serie temporal con un nivel de significancia del 5%

**5. INTERPRETACIÓN DE ACF Y PACF - Transformación de la serie temporal**

En la naturaleza es dificil encontrar una **serie naturalmente estacional**, por ello se hacen transformaciones a la misma y se convierte la serie en estacional al eliminar la tendencia y la estacionalidad; y poder posteriormente aplicar métodos de predicción. 

Del analisis anterior sabemos que nuestros datos necesitan ser transformados al mostrar tendencia a lo largo del tiempo, para ello utilizaremos la **diferenciación de orden 1.**

```python
st_difer_dia = num_sin_año.diff()
```

```python
plt.plot(st_difer_dia)
```

Una vez que los datos ya no muestran tendencia, vamos a representar las gráficas de autocorrelacion ACF y autocorrelación parcial PACF, de donde sacamos las siguientes conclusiones:

        La serie temporal tiene las siguientes correlaciones con los rezagos:
         - dia posteriores 1 y 2 
         - cada 7 dias (semanal) 


```python
st_difer_dia.dropna(inplace=True)
```

```python
fig,ax = plt.subplots(2,1,figsize=(15,8))
fig = sm.graphics.tsa.plot_acf(st_difer_dia, lags=100, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(st_difer_dia, lags=100, ax=ax[1])
ax[0].set_xlim(-1, 100)
ax[1].set_xlim(-1, 100)
plt.show()
```

Vamos a eliminar tambien el caracter estacional de la serie:

```python
st_difer_dia_sem = num_sin_año.diff().diff(7)
```

```python
plt.plot(st_difer_dia_sem)
```

```python
st_difer_dia_sem.dropna(inplace=True)
```

```python
fig,ax = plt.subplots(2,1,figsize=(15,8))
fig = sm.graphics.tsa.plot_acf(st_difer_dia_sem, lags=100, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(st_difer_dia_sem, lags=100, ax=ax[1])
ax[0].set_xlim(-1, 100)
ax[1].set_xlim(-1, 100)
plt.show()
```
